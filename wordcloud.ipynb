```python
import wordcloud
import wikipedia
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt


# 위키백과 사전의 컨텥츠 제목을 명시해 준다.
wiki = wikipedia.page('Artificial intelligence')
# 이 페이지의 텍스트 컨텐츠를 추출하도록 한다.
text = wiki.content
s_words = STOPWORDS.union({'one', 'using', 'first', 'two', 'make', 'use'})
wordcloud = WordCloud(background_color = 'white', width = 1200, height = 1000, stopwords= s_words).generate(text)
plt.figure(figsize=(10, 8)) # 화면에 이미지 크기 설정
plt.imshow(wordcloud)
plt.show()
```

```python
# 위키백과 사전의 컨텥츠 제목을 명시해 준다.
wiki2 = wikipedia.page('|Python|')
# 이 페이지의 텍스트 컨텐츠를 추출하도록 한다.
text = wiki2.content
s_words2 = STOPWORDS.union({'called', 'used', 'subset', 'based', 'eggs', 'may'})
wordcloud2 = WordCloud(background_color = 'black', width = 1200, height = 1000, stopwords= s_words2).generate(text)
plt.figure(figsize=(10, 8)) # 화면에 이미지 크기 설정
plt.imshow(wordcloud2)
plt.show()
```
