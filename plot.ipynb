```python
plt.plot(df['kor'], 'k--', marker='o', label='korean')
plt.plot(df['math'], marker='^', label='math')
plt.legend(loc = 'upper left') # 차트 범례, 왼쪽 위에 차트 레이블을 보여줌, loc = 'upper right' 오른쪽위로 'lower right' 아래 위쪽
plt.show()
```

```python
x = [6,2,1,1,4,4,6,4,7,7]
y = [5,4,2,9,9,0,10,3,1,2]

plt.plot(x, y, 'ms') #'bo' b는 색상 o는 모양, m* 마젠타에 *모양 h는 헥사 6각형
plt.show()
```

```python
x = ['A', 'B', 'C']
y = [100, 150, 120]

plt.bar(x,y, width=0.5, color = 'orange')
plt.title('Test Graph')
plt.xlabel('Group')
plt.ylabel('Score')
plt.show()
```
