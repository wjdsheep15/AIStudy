```python
import matplotlib.pyplot as plt
import seaborn as sns

tips_data = sns.load_dataset('tips')
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='day', data=tips_data)
plt.title('Total bill & tips')
plt.show()
```

```python
sns.countplot(x='day', data = tips_data)
plt.show()
```

```python
sns.pairplot(tips_data, hue='time')
plt.show()
```

```python
df = tips_data.corr()
sns.heatmap(data=df, annot=True, fmt='.2f', cbar=False)
plt.show()
```
