```python
from konlpy.tag import Okt
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

t = Okt()

text = ""
with open("kakao.txt", "r", encoding="utf-8" ) as f :
  lines = f.readlines()
  for line in lines:
    if '] [' in line :
      text+=line.split('] ')[2].replace('ㅋ', '').replace('ㅠ','').replace('ㅜ','').replace('사진\n', '').replace(' 이모티콘\n','').replace(' 삭제된 메시지입니다', '')
tokens_ko = t.nouns(text) #명사로 나눔

print(tokens_ko)
s_words = STOPWORDS.union({'ㅇㅇ', '승근', '아', '나', '지금', '오늘', '근데', '그냥', '진짜', '난', '어우', '창민이', '이모티콘', '얍얍', '승근씨', '배고파', 'ㄴㄴ', '됨', 'ㅅㅂ', '하기', '저','아니다', '뭐야', '한데',
                           '무슨', '했음', '맞다', '겁나', '탕수육', '뭐', '아니', '저', 'ㅇ', '이거', '일단', '윤카', '좀', '아님', '시발', '그럼', '다'})
path =  '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
wc = WordCloud(font_path = path, width = 2000, height = 1500, stopwords=s_words).generate(text)
plt.figure(figsize=(10, 8))
plt.imshow(wc)
plt.show()
```
